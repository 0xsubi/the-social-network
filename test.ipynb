```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
# Shift the timezone by -8 hours = 28800 seconds
df['timestamp'] = df['timestamp'] - 28800

# Convert to DateTime
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
```

```python
# Total number of data points
print(df.shape[0])
```

```python
# First timestamp
df['timestamp'].values.min()
```

```python
# Last timestamp
df['timestamp'].values.max()
```

```python
# Total number of posts created, posts viewed, logins, logouts
post_creates = df[df['action'] == 'MAKE_POST']
post_views = df[df['action'] == 'VIEW_POST']
logins = df[df['action'] == 'LOGIN']
unique_logins = logins.drop_duplicates(subset='user_id', keep="first")
logouts = df[df['action'] == 'LOGOUT']
```

```python
# Group data month wise
post_creates.set_index('timestamp',inplace=True)
post_views.set_index('timestamp',inplace=True)
logins.set_index('timestamp',inplace=True)
unique_logins.set_index('timestamp',inplace=True)
logouts.set_index('timestamp',inplace=True)

post_creates_by_month = post_creates.resample('M').count()
post_views_by_month = post_views.resample('M').count()
logins_by_month = logins.resample('M').count()
unique_logins_by_month = unique_logins.resample('M').count()
logouts_by_month = logouts.resample('M').count()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(post_creates_by_month.index.values,
        post_creates_by_month['action'],
        color='purple')
ax.set(xlabel="Month",
       ylabel="Posts Created",
       title="Posts Created over Months on Fakebook")
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(post_views_by_month.index.values,
        post_views_by_month['action'],
        color='purple')
ax.set(xlabel="Month",
       ylabel="Posts Viewed",
       title="Posts Viewed over Months on Fakebook")
```

```python
# Plot logins
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(logins_by_month.index.values,
        logins_by_month['action'],
        color='purple')
ax.set(xlabel="Month",
       ylabel="Logins",
       title="Logins over Months on Fakebook")
```

```python
# Plot unique logins
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(unique_logins_by_month.index.values,
        unique_logins_by_month['action'],
        color='purple')
ax.set(xlabel="Month",
       ylabel="Logins",
       title="Unique Logins over Months on Fakebook")
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(logouts_by_month.index.values,
        logouts_by_month['action'],
        color='purple')
ax.set(xlabel="Month",
       ylabel="Logouts",
       title="Logouts over Months on Fakebook")
```
